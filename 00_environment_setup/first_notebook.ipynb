```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 设置 Matplotlib 支持中文
plt.rcParams["font.sans-serif"] = ["Microsoft YaHei"]  # 微软雅黑，Windows 一般都有
plt.rcParams["axes.unicode_minus"] = False             # 正常显示负号
```

```python
data = {
    "姓名": ["小明", "小红", "小刚", "小李"],
    "数学": [85, 92, 78, 88],
    "英语": [90, 86, 80, 95],
    "编程": [88, 91, 84, 89]
}

df = pd.DataFrame(data)
df
```

```python
df["平均分"] = df[["数学", "英语", "编程"]].mean(axis=1)
df
```

```python
df.plot(x="姓名", y="平均分", kind="bar", title="学生平均分")
plt.show()
```
